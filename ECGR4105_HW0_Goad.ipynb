```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Wgoad/ECGR4105/main/D3.csv')
df.head()
M=len(df)
M          # test to check if file is read properly
```

```python
# function definitions for cost and gradient descent

def compute_cost(X,y,theta):
    predictions = X.dot(theta) 
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J

def gradient_descent(X, y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors);
        theta = theta - sum_delta;
        cost_history[i] = compute_cost(X, y, theta)
        
    return theta, cost_history
```

```python
# X1 Regression
X1 = df.values[:, 0] # takes input values from column 1
y = df.values[:, 3] # gets output values from column 4
m = len(y) # number of training examples
```

```python
plt.scatter(X1,y, color='red',marker= '+')
plt.grid()
plt.rcParams["figure.figsize"] = (10,6)
plt.title('Problem 1 X1 Scatter Plot')
```

```python
# creating a matrix with a single column of ones
X_0 = np.ones((m,1))

# reshaping x into a 2D array of dimension m * 1
X_1 = X1.reshape(m,1)

X1 = np.hstack((X_0, X_1))
```

```python
# parameter settings for gradient descent
theta = [0., 0.]
iterations = 1500;
alpha = 0.1;

# calculated results from the gradient descent
theta, cost_history = gradient_descent(X1, y, theta, alpha, iterations)
cost_1 = compute_cost(X1, y, theta)
cost_1
```

```python
plt.scatter(X1[:,1], y, color='red', marker= '+', label= 'Training Data')
plt.plot(X1[:,1],X1.dot(theta), color='green', label='Linear Regression')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.title('Problem 1 Linear Regression Fit for X1')
plt.legend()
```

```python
plt.plot(range(1, iterations + 1),cost_history, color='blue')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Cost for X1')
```
